### Step 1. Import the necessary libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv')
tips
```

### Step 4. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']
```

### Step 5. Plot the total_bill column histogram

```python
plt.plot(tips.total_bill)
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
tips = sns.load_dataset('tips')
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', s = tips['size']*50, alpha = 0.5).set(title = 'Relationship between Total Bill, Tip and Size')
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.violinplot(x = tips['day'], y = tips['total_bill'], data = tips)
plt.xlabel('data')
plt.ylabel('Total Bill')
plt.title('relationship between days and total_bill value')
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(data=tips, x = 'tip', y = 'day', hue = 'size', s = tips['size']*50, alpha = 0.5)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x = tips['day'], y = 'total_bill', hue = 'time', data = tips)
plt.xlabel('Day')
plt.ylabel('total_bill')
plt.title('Total Bill by Day and Time')
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
fig,(ax1, ax2) = plt.subplots(1,2,figsize = (10,5))
# tip for lunch
ax1.hist(tips[tips['time']=='Lunch']['tip'], bins = 10, color = 'pink')
ax1.set_xlabel('Tip')
ax1.set_title('Tip for Lunch')
# tip for dinner
ax2.hist(tips[tips['time']=='Dinner']['tip'], bins=10, color = 'purple')
ax2.set_xlabel('Tip')
ax2.set_title('Tip for Dinner')
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (10,5))
# male or female smoker

ax1.scatter(tips[(tips['sex'] == 'Male') & (tips['smoker'] == 'Yes')]['total_bill'],
            tips[(tips['sex'] == 'Male') & (tips['smoker'] == 'Yes')]['tip'],
            label='Male Smoker', alpha = 0.5)
ax1.scatter(tips[(tips['sex'] == 'Female') & (tips['smoker'] == 'Yes')]['total_bill'],
            tips[(tips['sex'] == 'Female') & (tips['smoker'] == 'Yes')]['tip'],
            label='Female Smoker', alpha = 0.5)

ax1.set_xlabel('Total Bill')
ax1.set_ylabel('Tip')
ax1.set_title('Total Bill and Tip for Smokers')
ax1.legend()

# male or female non smoker

ax2.scatter(tips[(tips['sex'] == 'Male') & (tips['smoker'] == 'No')]['total_bill'],
            tips[(tips['sex'] == 'Male') & (tips['smoker'] == 'No')]['tip'],
            label='Male Smoker', alpha = 0.5)
ax2.scatter(tips[(tips['sex'] == 'Female') & (tips['smoker'] == 'No')]['total_bill'],
            tips[(tips['sex'] == 'Female') & (tips['smoker'] == 'No')]['tip'],
            label='Female Smoker', alpha = 0.5)

ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Tip')
ax2.set_title('Total Bill and Tip for Non Smokers')
ax2.legend()
```
